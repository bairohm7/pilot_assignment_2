features which are sex,smoke and region are not supported on the heatmap because they are the categorical values. A computer can not understand those categorical values, so we must decode them in order the computer to understand them. There are different methods/ways to deal with those categorical values. One of the method to deal with categorical values is called One-Hot-Encoding(Dummy Variables). This method can be implemented by either using a get_dummies() function which is found in pandas library or that implemented in scikit library.

```python
#Implementation of One-Hot-Encoding method
data_dummies = pd.get_dummies(df5)
display(data_dummies)

```

```python
print("New shape of the dataset is: {}".format(data_dummies.shape))
```

# splitting data into training and test set

```python
#Spliting the dataset
x = features = data_dummies.drop("charges", axis = 1)
y = data_dummies.iloc[:, 3]
print("x/features are:")
display(features)
print("y values are:\n{}".format(y))
```

```python
#Spliting the train set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)
```

# build the model to fit the training data

```python
#Building a linear regression model for multiple features
xmas = LinearRegression()
xmas.fit(x_train, y_train)
```

# evaluating the model performance

```python
#Evaluating the models
print("Train accuracy score: {}".format(xmas.score(x_train, y_train)))
print("Test accuracy score: {}".format(xmas.score(x_test, y_test)))
```

```python
#Try to use other metrics for evaluating a model
y_pred = xmas.predict(x_test)
print("Mean squred error: {}".format(mean_squared_error(y_test, y_pred)))
print("R2 score: {}".format(r2_score(y_test,y_pred)))
mean_squared_before = mean_squared_error(y_test, y_pred)
r2_score_before = r2_score(y_test, y_pred)
```

```python
#Model improvement
#Let us improve our model now
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

#Let us fit the scaler
scaler.fit(x_train)
#Then Transform the x_train set
x_train_scaled = scaler.transform(x_train)
print("Transformed shape: {}".format(x_train_scaled.shape))
print("Shape before transformation: {}".format(x_train.shape))
```

The MinMaxScaler, shifts the data such that all features are exactly between 0 and 1

```python
#Then we transform the x_test set
x_test_scaled = scaler.transform(x_test)
```

```python
#Then we build a model using a scaled data
lr11 = LinearRegression()
lr11.fit(x_train_scaled, y_train)

#now let's evaluate a model

y_pred = lr11.predict(x_test_scaled)
print("A new R2 score is: {}".format(r2_score(y_test, y_pred)))
print("A new mean squared error is: {}".format(mean_squared_error(y_test, y_pred)))
mean_squared_after = mean_squared_error(y_test, y_pred)
r2_score_after = r2_score(y_test, y_pred)
```

```python
#Comparison of the values before and after scaling the data
print("Mean Squared error before scaling is: {}\nMean Squared error after scaling is: {}".format(mean_squared_before, mean_squared_after))
print("R2 score before scaling is: {}\nR2 score after scaling is: {}".format(r2_score_before, r2_score_after))
```

After scaling the data, there is small decrease in mean squared error. Also there is small increase of r2 score.
So we may conclude that;
R2_score is inversely proportional to the mean squared error
